```python


def calculate_expected_charges(row):
    shipment_type_mapping = {
    "Forward Charges": "fwd",
    "Forward and RTO Charges": "fwd_and_rto"
    }

    # Map the shipment type
    mapped_type = shipment_type_mapping.get(row['shipment_type'], None)
    print(mapped_type)
    if not mapped_type:
        return np.nan  # If no valid mapping, return NaN

    # Handle the case for "Forward and RTO Charges"
    if mapped_type == "fwd_and_rto":
        # Calculate forward charges
        forward_fixed = rate_card_df[f"fwd_{row['Zone'].lower()}_fixed"].values[0]
        forward_additional = rate_card_df[f"fwd_{row['Zone'].lower()}_additional"].values[0]
        forward_excess_weight = max(0, row['Weight Slab'] - 0.5)
        forward_charge = forward_fixed + (forward_excess_weight / 0.5) * forward_additional

        # Calculate RTO charges
        rto_fixed = rate_card_df[f"rto_{row['Zone'].lower()}_fixed"].values[0]
        rto_additional = rate_card_df[f"rto_{row['Zone'].lower()}_additional"].values[0]
        rto_excess_weight = max(0, row['Weight Slab'] - 0.5)
        rto_charge = rto_fixed + (rto_excess_weight / 0.5) * rto_additional

        # Total charge is the sum of forward and RTO charges
        return forward_charge + rto_charge
    else:
        # Calculate for forward-only ("fwd")
        fixed_column = f"{mapped_type}_{row['Zone'].lower()}_fixed"
        additional_column = f"{mapped_type}_{row['Zone'].lower()}_additional"

        # Check if the columns exist
        if fixed_column not in rate_card_df.columns or additional_column not in rate_card_df.columns:
            return np.nan

        # Extract charges
        fixed_charge = rate_card_df.loc[rate_card_df['zone'] == zone, fixed_column].values[0]
        additional_charge = rate_card_df.loc[rate_card_df['zone'] == zone, additional_column].values[0]


        # Calculate excess weight
        excess_weight = max(0, row['Weight Slab'] - 0.5)

        # Total charge calculation
        return fixed_charge + (excess_weight / 0.5) * additional_charge
```
